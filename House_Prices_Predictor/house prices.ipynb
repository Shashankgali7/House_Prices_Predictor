```python
numerical_column.loc[numerical_column['Age_house']<0,'YrSold']=2009
numerical_column['Age_house']=(numerical_column['YrSold']-numerical_column['YearBuilt'])
numerical_column['Age_house'].describe()
```

```python
numerical_column['TotalBsmtBath']=numerical_column['BsmtFullBath']+numerical_column['BsmtHalfBath']*0.5
numerical_column['TotalBath']=numerical_column['FullBath']+numerical_column['HalfBath']*0.5
```

```python
numerical_column['TotalSF']=numerical_column['TotalBsmtSF']+numerical_column['1stFlrSF']+numerical_column['2ndFlrSF']
```

```python
numerical_column.head()
```

```python
bin_map  = {'TA':2,'Gd':3, 'Fa':1,'Ex':4,'Po':1,'None':0,'Y':1,'N':0,'Reg':3,'IR1':2,'IR2':1,'IR3':0,"None" : 0,
            "No" : 2, "Mn" : 2, "Av": 3,"Gd" : 4,"Unf" : 1, "LwQ": 2, "Rec" : 3,"BLQ" : 4, "ALQ" : 5, "GLQ" : 6}
```

```python
objects_column['ExterQual'] = objects_column['ExterQual'].map(bin_map)
objects_column['ExterCond'] = objects_column['ExterCond'].map(bin_map)
objects_column['BsmtCond'] = objects_column['BsmtCond'].map(bin_map)
objects_column['BsmtQual'] = objects_column['BsmtQual'].map(bin_map)
objects_column['HeatingQC'] = objects_column['HeatingQC'].map(bin_map)
objects_column['KitchenQual'] = objects_column['KitchenQual'].map(bin_map)
objects_column['FireplaceQu'] = objects_column['FireplaceQu'].map(bin_map)
objects_column['GarageQual'] = objects_column['GarageQual'].map(bin_map)
objects_column['GarageCond'] = objects_column['GarageCond'].map(bin_map)
objects_column['CentralAir'] = objects_column['CentralAir'].map(bin_map)
objects_column['LotShape'] = objects_column['LotShape'].map(bin_map)
objects_column['BsmtExposure'] = objects_column['BsmtExposure'].map(bin_map)
objects_column['BsmtFinType1'] = objects_column['BsmtFinType1'].map(bin_map)
objects_column['BsmtFinType2'] = objects_column['BsmtFinType2'].map(bin_map)
PavedDrive =   {"N" : 0, "P" : 1, "Y" : 2}
objects_column['PavedDrive'] = objects_column['PavedDrive'].map(PavedDrive)
```

```python
rest_object_column = objects_column.select_dtypes(include=['object'])
objects_column=pd.get_dummies(objects_column,columns=rest_object_column.columns)
```

```python
df_final=pd.concat([objects_column,numerical_column],axis=1,sort=False)
df_final.head()
```

```python
df_final = df_final.drop(['Id',],axis=1)

df_train = df_final[df_final['c_train'] == 1]
df_train = df_train.drop(['c_train',],axis=1)


df_test = df_final[df_final['c_train'] == 0]
df_test = df_test.drop(['SalePrice'],axis=1)
df_test = df_test.drop(['c_train',],axis=1)
```
